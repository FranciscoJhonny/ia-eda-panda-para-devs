## Transformação de dados

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 1 com astype -> Essa abordagem vai dar erro porque não pode converte 
# pois tem espaço em uma coluna ou mais dados
df_contracts.TotalCharges = df_contracts.TotalChages.astype(float)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 2 com to_numeric -> Essa abordagem vai dar erro porque não pode converte 
# pois tem espaço em uma coluna ou mais dados
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 3 com to_numeric com coerce -> se encontra erros algum coluna com espaço vazil ele transforma em null
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges, errors='coerce')
```

```python
# Mostrar a estrutura / schema do Dataframe
df_contracts.info()
```

## Unificar Dateframe de Customers, Services e Contracts

### Contar quantidade de registros por dataframs

```python
len(df_customers)
```

```python
len(df_contracts)
```

```python
len(df_services)
```

```python
#Renomear colunas
df_services.rename(columns={'customerID':'IDCliente'},inplace=True)
```

```python
df_services.info()
```

```python
# Unificar Dataframe de Customers com Services, criando um terceiro DataFrame
df_temp = df_customers.merge(df_services,on=['IDCliente'])
```

```python
df_temp.info()
```

```python
df_temp.head(5)
```

```python
# Unificado df_temp com Contracts, usando colunas de função com nomes distintos
df_churn_temp = df_temp.merge(df_contracts,left_on=['IDCliente'],right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Unificar os três dataframes ao mesmo tempo, com colunas com nomes diferentes
df_churn = df_customers.merge(df_services, on=['IDCliente']).merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
# Removendo coluna de um Dataframe
df_churn.drop(['customerID'], axis=1,inplace=True)
```

```python
df_churn.info()
```
